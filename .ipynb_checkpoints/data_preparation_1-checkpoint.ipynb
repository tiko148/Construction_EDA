```python
import pandas as pd

```

```python
'''Setting option for pandas to show maximum rows and columns'''

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
'''Reading data for all houses sold in North Eeast from 2017- 2021'''
North_East = pd.read_excel("New_Houses_Sold_region_SalesPrice.xls",sheet_name="QtrSalGrp Reg Pct",header=None, names=["Period", "Total", "Under 200", "200-299","300-499",
                                                                 "500-749","750 and Over",],skiprows=69, nrows= 20, 
                                                                  index_col = None, usecols="A:G")
North_East['Region'] = 'North East'
North_East.set_index('Period', inplace =True)
North_East
#[Thousands of houses.  Components may not add to total because of rounding].  


```

```python
'''Reading data for all houses sold in Mid West from 2017- 2021'''
Mid_West = pd.read_excel("New_Houses_Sold_region_SalesPrice.xls",sheet_name="QtrSalGrp Reg Pct",header=None, names=["Period", "Total", "Under 150", "150-199","200-299",
                                                                 "300-499","500 and Over",],skiprows=69, nrows= 20, 
                                                                  index_col = None, usecols="A,H:M")

Mid_West['Region'] = 'Mid West'
Mid_West.set_index("Period", inplace = True)
Mid_West
```

```python
'''Reading data for all houses sold in South from 2017- 2021'''
South  = pd.read_excel("New_Houses_Sold_region_SalesPrice.xls",sheet_name="QtrSalGrp Reg Pct",header=None, names=["Period", "Total", "Under 150", "150-199","200-299",
                                                                 "300-499","500 and Over",],skiprows=69, nrows= 20, 
                                                                  index_col = None, usecols="N:T")

South['Region'] = 'South'
South.set_index('Period', inplace = True)
South
```

```python
'''Reading data for all houses sold in West from 2017- 2021'''

West = pd.read_excel("New_Houses_Sold_region_SalesPrice.xls",sheet_name="QtrSalGrp Reg Pct",header=None, names=["Period", "Total", "Under 200", "200-299","300-499",
                                                                 "500-749","750 and Over",],skiprows=69, nrows= 20, 
                                                                  index_col = None, usecols="A, U:Z")
West['Region'] = 'West'
West.set_index("Period", inplace = True)
West
```

```python
new_houses_sold_price_by_region = pd.concat([North_East, West, South, Mid_West])
#new_houses_sold_price_by_region.to_csv('new_price_by_region')
new_houses_sold_price_by_region
```
