```python
'''
Clean excel file called 'houses_sold_type_of_financing.xls' that outlines type of financing for home purchases
in the United States. The data is divided into quarters.

'''



sold_finance_type = pd.read_excel('houses_sold_type_of_financing.xls',sheet_name="Sold Finance Qtr",header=None, names=["Period","Total Sold","Conventional",
                                                                                                 "FHA Insured", "VA Guaranteed", "Cash"], 
                                                                 skiprows=110, nrows= 32, 
                                                                  index_col = None, usecols="A:E,G",)
sold_finance_type.set_index('Period', inplace = True)
sold_finance_type



'''This if for 2014- 2021 --> differnt period than the other ones'''


for i in sold_finance_type.columns:
    sold_finance_type[i + ' Rate of Change'] = sold_finance_type[[i]].pct_change()


new_oder = list(sold_finance_type.columns)

new_order = ['Total Sold',
            'Total Sold Rate of Change',
 'Conventional',
 'Conventional Rate of Change',
 'FHA Insured',
 'FHA Insured Rate of Change',
 'VA Guaranteed',
'VA Guaranteed Rate of Change',
 'Cash',
'Cash Rate of Change']

sold_finance_type = sold_finance_type[new_order]


sold_finance_type.to_csv('sold_finance_type.csv')

sold_finance_type
```

```python
'''
Get hoomes sold data for the United States (Similar to the first graph, but for USA) in the thousands.


'''



houses_sold_by_sales_price_USA= pd.read_excel("houses_sold_by_sales_price.xls",sheet_name="QtrSalGrp US",header=None, names=["Period","USA","0-125",
                                                                                                 "125-149", "150-199", "200-249",
                                                                               '250-299','300-399','400-499','500-749','750 and Above'], 
                                                                 skiprows=56, nrows= 32, 
                                                                  index_col = None, usecols="A:K")


houses_sold_by_sales_price_USA.set_index('Period', inplace = True)
houses_sold_by_sales_price_USA

for i in houses_sold_by_sales_price_USA.columns:
    houses_sold_by_sales_price_USA[i + ' Rate of Change'] = houses_sold_by_sales_price_USA[[i]].pct_change()

    

new_cols = list(houses_sold_by_sales_price_USA.columns)
new_cols
new_cols = ['USA',
   'USA Rate of Change',
 '0-125',
 '0-125 Rate of Change',
 '125-149',
  '125-149 Rate of Change',
 '150-199',
  '150-199 Rate of Change',            
 '200-249',
  '200-249 Rate of Change',
 '250-299',
 '250-299 Rate of Change',
 '300-399',
 '300-399 Rate of Change',
 '400-499',
  '400-499 Rate of Change',
 '500-749',
  '500-749 Rate of Change',
 '750 and Above',
 '750 and Above Rate of Change']


 
houses_sold_by_sales_price_USA = houses_sold_by_sales_price_USA[new_cols]
 

houses_sold_by_sales_price_USA.to_csv('houses_sold_by_sales_price_USA.csv')

houses_sold_by_sales_price_USA

```
