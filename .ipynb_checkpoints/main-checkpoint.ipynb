```python
fig = go.Figure()


fig.add_trace(go.Bar(x = auth_not_started.Date, y = auth_not_started['United States'], name = 'USA',marker_color='gray'))

fig.add_trace(go.Bar(x = auth_not_started.Date, y = auth_not_started['North East'], name = 'Northeast',marker_color='red'))
fig.add_trace(go.Scatter(x = auth_not_started.Date, y = auth_not_started['North East'], name = 'Northeast-line', mode = 'lines'))
fig.add_trace(go.Bar(x = auth_not_started.Date, y = auth_not_started['West'], name = 'West',))
fig.add_trace(go.Scatter(x = auth_not_started.Date, y = auth_not_started['West'],name = 'West-line',mode='lines',))

fig.add_trace(go.Bar(x = auth_not_started.Date, y = auth_not_started['South'], name = 'South',marker_color = '#054907'))
fig.add_trace(go.Scatter(x = auth_not_started.Date, y = auth_not_started['South'], name = 'South-line',mode = 'lines'))
fig.add_trace(go.Bar(x = auth_not_started.Date, y = auth_not_started['Mid West'], name = 'Midwest',marker_color = '#000080'))
fig.add_trace(go.Scatter(x = auth_not_started.Date, y = auth_not_started['Mid West'], name = 'Midwest-line',mode = 'lines'))

fig.update_layout(title_text = 'Permits Authorized but Not Started',
                  
                  yaxis_title = 'Not Started in Thousands',
                  yaxis_range = [0,200],
                  yaxis_dtick = 10,
                  xaxis_title = 'Period By Quarter',
                  font = dict(family = 'verdana',
                             size = 12,
                             color = 'black',
                             ),
                  yaxis_showgrid = False,
                  yaxis_linecolor = 'white',
                  legend = dict(bgcolor ='white',
                                title = 'Legend',
                                font = dict(color = 'black',size = 8)
                               ),
                  plot_bgcolor = 'white',
                  clickmode = 'select',
                  dragmode = 'drawline',
                  transition = dict(duration = 1000,
                                   easing = 'bounce-in',
                                   ordering = 'traces first'))
                  
                  
fig.add_vrect(x0 = '2020-12-1', x1 = '2021-12-30', col =1, fillcolor="gray",opacity=0.15, line_width=0)

fig.show()
                  
                  
                  
                  
                  
                  
```
